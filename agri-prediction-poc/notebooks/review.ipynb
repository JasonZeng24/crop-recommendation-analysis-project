# Pandas and Numpy

```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/Crop_recommendation.csv')

print(df.shape)

df.head()
```

### Filtering

```python
high_rain_crops = df[df["rainfall"]> 200]

high_raincrops_list = high_rain_crops["label"].unique()

print(high_raincrops_list)
```

```python
extreme_weather = df[(df["temperature"]>30) & (df["humidity"]>90)]

print(extreme_weather['label'].unique())
print(len(extreme_weather))
extreme_weather.head()
```

### Statistics & Masking

```python
df['is_hot'] = np.where(df['temperature'] > 25, 1, 0)

display(df.head())

matrix_np = df.iloc[:,0:3].values

print(matrix_np)

print(np.mean(matrix_np, axis = 0))
```

### GroupBy

```python
crop_summary = df.groupby('label').mean()

crop_summary = crop_summary.sort_values(by = 'rainfall', ascending = False)

display(crop_summary.head())
```
